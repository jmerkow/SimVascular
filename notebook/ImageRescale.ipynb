```python
import SimpleITK as sitk
import numpy as np
import sys,os
%matplotlib inline
```

```python
imgfn='/home/jameson/Data/I2IResults/3DImages_Mnms/0087_1001_I2I.mha'
img=sitk.ReadImage(imgfn)
```

```python
ndimg=sitk.GetArrayFromImage(img)
ndimg.min(),ndimg.max(),ndimg.mean(),ndimg.var()
```

```python
outdir='/home/jameson/Data/I2IResults/3DImages_Mnms-rescale/'
outfn = os.path.join(outdir,os.path.split(imgfn)[1])
if not os.path.exists(outdir): os.makedirs(outdir)
print(outfn)
```

```python
img*=255
caster=sitk.CastImageFilter()
caster.SetOutputPixelType(sitk.sitkInt16)
img=caster.Execute(img)

sitk.WriteImage(img,outfn)
```

```python
ndimg=sitk.GetArrayFromImage(img)
ndimg.min(),ndimg.max(),ndimg.mean(),ndimg.var()
```
